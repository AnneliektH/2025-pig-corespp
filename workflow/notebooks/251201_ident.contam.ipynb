## Get the single copy marker genes of species. 
- Some genomes were removed due to Artydactola contamination
- Get list of these names and exclude them / put in separate folder
- Comes with renaming, shouldnt be too hard. 

```python
import pandas as pd
```

```python
df = pd.read_csv('/home/ctbrown/scratch3/sourmash-midgie-raker/outputs.ath/host/bins.x.host.manysearch.csv')
len(df)
```

```python
df = df[["query_name"]]
df[["new_name", "old_name"]] = df["query_name"].str.split("(", n=1, expand=True)
# split 'column_name' on spaces into two new columns
df[["ident", "tax"]] = df["new_name"].str.split(" ", n=1, expand=True)
df = df[["ident"]]
```

```python
df2 = pd.read_csv('/home/ctbrown/scratch3/sourmash-midgie-raker/outputs.ath/host/gtdb.x.host.manysearch.csv')
```

```python
# split 'column_name' on spaces into two new columns
df2[["ident", "tax"]] = df2["query_name"].str.split(" ", n=1, expand=True)
df2 = df2[["ident"]]
```

```python
df_combined = pd.concat([df, df2], ignore_index=True)

# drop duplicates
df_combined = df_combined.drop_duplicates()
```

```python
df_combined.to_csv('/group/ctbrowngrp2/amhorst/2025-pig-corespp/results/csv_files/251201_ident.contaminated.csv', index=False)
```
