```python
import pandas as pd
import glob
```

```python
n_samples = 100
df = pd.read_csv("/group/ctbrowngrp2/amhorst/2025-pig-corespp/resources/SRA_meta.3216.csv")
df2 = pd.read_csv('/group/ctbrowngrp2/amhorst/2025-pig-corespp/resources/assembled.txt')
```

```python
# subset the bioprojects to max 150 samples each
df = (
    df
    .drop_duplicates(["bioproject", "sample_name"])  # 1. remove duplicate biosamples per bioproject
    .groupby("bioproject")
    .apply(lambda x: x.sample(n=150, replace=False) if len(x) > 150 else x)
    .reset_index(drop=True)
)
```

```python
df['bioproject'].value_counts()
```

```python
# if only assembled samples
df1 = df[df["acc"].isin(df2["acc"])]
```

```python
df_subset = (
    df.sample(frac=1, random_state=42)      # shuffle whole dataframe
      .drop_duplicates("biosample")       # keep only one row per sample_name
      .sample(n=n_samples, random_state=42)         # pick x rows
      .reset_index(drop=True)
)
```

```python
df_subset1 = (
    df1.sample(frac=1, random_state=42)      # shuffle whole dataframe
      .drop_duplicates("biosample")       # keep only one row per sample_name
      .sample(n=n_samples, random_state=42)         # pick x rows
      .reset_index(drop=True)
)
```

```python
df_subset1['acc'].to_csv('/group/ctbrowngrp2/amhorst/2025-pigparadigm/results/rand_subset.assembled.100.txt', index=False, header=False)
```

```python
df_subset['acc'].to_csv('/group/ctbrowngrp2/amhorst/2025-pigparadigm/results/rand_subset.3216.100.txt', index=False, header=False)
```
