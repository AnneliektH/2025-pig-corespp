## Select isolate genomes from NCBI and only use these
- 6 of our species have isolates available 

Went ot NCBI genomes, searched for species, then excluded MAGs,  atypical genomes, Genomes from large multi-isolate projects.
im doing this manually because thats easier

```python
import pandas as pd
from pathlib import Path
```

```python

names = [
    "Lactobacillus_amylovorus",
    "Roseburia_inulinivorans",
    "Mogibacterium_A_kristiansenii",
    "Gemmiger_qucibialis",
    "Holdemanella_porci",
    "Phascolarctobacterium_A_succinatutens"
]
```

```python

base = Path("/group/ctbrowngrp2/amhorst/2025-pig-corespp/results/pangenome")
outdir = Path("/group/ctbrowngrp2/amhorst/2025-pig-corespp/results/isolate_genomes")

tsv = pd.read_csv("/group/ctbrowngrp2/amhorst/2025-pig-corespp/results/isolate_genomes/genome_isolates.tsv", sep="\t")
acc = set(tsv["Assembly Accession"])

for name in names:
    csv_file = base / name / f"{name}xgtdb.csv"
    df = pd.read_csv(csv_file)
    out = df[df["ident"].isin(acc)]
    out.to_csv(outdir / f"{name}_filtered.csv", index=False)
```
